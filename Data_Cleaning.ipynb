```python
onehot_columns = ['sex_cd', 'rucc_category','cms_race_cd','cons_mobplus', 'cms_ra_factor_type_cd', 'cons_homstat', 'lang_spoken_cd','cms_orig_reas_entitle_cd']
onehot_df = df1[onehot_columns]
onehot_df = pd.get_dummies(onehot_df, columns = onehot_columns)
score_onehot_drop = df1.drop(onehot_columns, axis = 1)
score_onehot = pd.concat([score_onehot_drop, onehot_df], axis = 1)
```
